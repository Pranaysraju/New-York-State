```python


sns.set(rc={'figure.figsize':(12,10)})
sns.set(style="white", context="talk")

history = pd.read_sql_query('SELECT LAW_CAT_CD,CMPLNT_FR_DT,CMPLNT_FR_TM,OFNS_DESC FROM tbl_nypdHistory',con)

df2 = history.copy()
def remove_dates(x):
    if x[2] > '2050':
        x = None
    elif x[2] < '2010':
        x = None
    else:
        aa= '/'.join(x)
        return (aa)

df2 = df2.join(df2['LAW_CAT_CD'].str.get_dummies())
df2.dropna(subset=['CMPLNT_FR_DT'], inplace=True)
df2['CMPLNT_FR_DT'] = df2['CMPLNT_FR_DT'].str.split("/")
df2['CMPLNT_FR_DT'] = df2['CMPLNT_FR_DT'].apply(lambda x: remove_dates(x))

df2['Date_OCCRD'] = df2['CMPLNT_FR_DT'] +' '+df2['CMPLNT_FR_TM'] 
df2['Date_OCCRD'] = pd.to_datetime(df2['Date_OCCRD'])
df2.set_index('Date_OCCRD', inplace=True)   

```

```python
dfYear = df2.iloc[:900000] 
dfYear.index = dfYear.index.year
vis0 = dfYear.groupby([dfYear.index.get_level_values(0)])[['VIOLATION','FELONY','MISDEMEANOR']].sum()
vis0.plot.bar()
plt.title('Amount of crime per year')
plt.xlabel('Year')
plt.show()

dfMonth = df2.iloc[:900000] 
dfMonth.index = dfMonth.index.month
vis1 = dfMonth.groupby([dfMonth.index.get_level_values(0)])[['VIOLATION','FELONY','MISDEMEANOR']].sum()
vis1.plot.bar()
plt.title('Amount of crime per month')
plt.legend(loc='upper left')
plt.xlabel('Month')
plt.show()

dfDay = df2.iloc[:900000] 
print ("Significantly more crimes happen on the 1st of the month")
dfDay.index = dfDay.index.day
vis2 = dfDay.groupby([dfDay.index.get_level_values(0)])[['VIOLATION','FELONY','MISDEMEANOR']].sum()
vis2.plot.bar(figsize=(12,6))
plt.title('Amount of crime per day')
plt.xlabel('Day of Month')
plt.show()

dfHour = df2.iloc[:900000] 
print ("It seems like more crimes happen in the afternoon than at night")
dfHour.index = dfHour.index.hour  
vis3 = dfHour.groupby([dfHour.index.get_level_values(0)])[['VIOLATION','FELONY','MISDEMEANOR']].sum()
vis3.plot.bar(figsize=(12,6))
plt.title('Amount of crime per hour')
plt.xlabel('Time of day')
plt.show()
```

```python
con = sqlite3.connect('ny.db')
df=pd.read_sql_query("SELECT substr(CMPLNT_FR_DT,7,8) as Year ,CMPLNT_NUM,BORO_NM FROM NYPD where substr(CMPLNT_FR_DT,7,8) in ('2006','2007','2008','2009','2010','2012','2013','2015','2016') ",con)
boroughs = set(df[df.BORO_NM.notnull()].BORO_NM);
df2=df.groupby(['Year'])[['CMPLNT_NUM']].count().reset_index();
df2['Year'] = df['Year'].astype(int)
df2.tail(10)

df1=pd.read_sql_query("SELECT substr(CMPLNT_FR_DT,7,8) as Year ,CMPLNT_NUM,BORO_NM FROM NYPD where substr(CMPLNT_FR_DT,7,8) in ('2010','2011','2012','2013','2015','2016','2017','2018','2019') ",con)

df3=df1.groupby(['Year','BORO_NM'])[['CMPLNT_NUM']].count().reset_index().tail(90)
df3['Year'] = df['Year'].astype(int)
```

```python
def valfunc(row,val_dict,column):
    
    return val_dict[row[column]]
```

```python
from scipy import stats
slope, intercept, rvalue, pvalue, stderr = stats.linregress(df2.Year, df2.CMPLNT_NUM)
g=sns.lmplot(x='Year',y='CMPLNT_NUM',data=df2,palette='YlGnBu');
g=(g.set_axis_labels("Year","Number of Offences"));
plt.title("Crime Rate in NYC, correlation = {:.2f}, R2 = {:.2f}, p-value = {:.4f}".format(rvalue,rvalue**2,pvalue));
```
